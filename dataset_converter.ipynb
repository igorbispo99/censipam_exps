```python
import geopandas
```

```python
df = geopandas.read_file("data/train_labels.geojson")
```

```python
classes = ['18', '73', '89', '25', '24', '23', '19', '21', '17', '35']

df['type_id'] = df['type_id'].apply(str)
df = df[df['type_id'].isin(classes)]
```

```python
!mkdir data/splitted_json
```

```python
%%capture

for img_name in df.image_id.unique():
    if img_name != '1395.tif': # For some reason, this tif file is missing
        sub_df = df[df.image_id == img_name]
        
        sub_df.to_file("data/splitted_json/" + img_name[:img_name.find('.')] + '.geojson', driver="GeoJSON")
```
